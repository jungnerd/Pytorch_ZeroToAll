# XOR 문제 - 단층 퍼셉트론 구하기

```python
import torch
import torch.nn as nn
```

```python
device = torch.device("mps")
torch.cuda.manual_seed(777)
```

```python
X = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]]).to(device)
Y = torch.FloatTensor([[0], [1], [1], [0]]).to(device) # XOR 출력값
```

```python
linear = nn.Linear(2, 1, bias=True)
sigmoid = nn.Sigmoid()
model = nn.Sequential(linear, sigmoid).to(device)
```

```python
# 비용 함수와 옵티마이저 정의
criterion = nn.BCELoss().to(device)
optimizer = torch.optim.SGD(model.parameters(), lr=1)
```

```python
for step in range(10001):
    optimizer.zero_grad()
    hypothesis = model(X)
    
    # 비용 함수
    cost = criterion(hypothesis, Y)
    cost.backward()
    optimizer.step()
    
    if step % 100 == 0:
        print(step, ":", cost.item())
```

```python
with torch.no_grad():
    hypothesis = model(X)
    predicted = (hypothesis > 0.5).float()
    accuracy = (predicted == Y).float().mean()
    print("모델의 출력값(Hypothesis): \n", hypothesis.detach().cpu().numpy())
    print("모델의 예측값(Predicted: \n", predicted.detach().cpu().numpy())
    print("실제값(Y): \n", Y.cpu().numpy())
    print("정확도(Accuracy): \n", accuracy.item())
```

# XOR 문제 - 다층 퍼셉트론 구하기

```python
import torch
import torch.nn as nn
```

```python
device = torch.device("mps")

torch.manual_seed(777)
```

```python
X = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]]).to(device) # XOR 입력값
Y = torch.FloatTensor([[0], [1], [1], [0]]).to(device) # XOR 출력값
```

```python
model = nn.Sequential(
    nn.Linear(2, 10, bias=True),
    nn.Sigmoid(),
    nn.Linear(10, 10, bias=True),
    nn.Sigmoid(),
    nn.Linear(10, 10, bias=True),
    nn.Sigmoid(),
    nn.Linear(10, 1, bias=True),
    nn.Sigmoid()
).to(device)
```

```python
criterion = nn.BCELoss().to(device)
optimizer = torch.optim.SGD(model.parameters(), lr=1)
```

```python
for epoch in range(10001):
    optimizer.zero_grad()
    # forward 연산
    hypothesis = model(X)
    
    # 비용 함수
    cost = criterion(hypothesis, Y)
    cost.backward()
    optimizer.step()
    
    if epoch % 100 == 0:
        print(epoch, cost.item())
```

```python
with torch.no_grad():
    hypothesis = model(X)
    predicted = (hypothesis > 0.5).float()
    accuracy = (predicted == Y).float().mean()
    print('모델의 출력값(Hypothesis): ', hypothesis.detach().cpu().numpy())
    print('모델의 예측값(Predicted): ', predicted.detach().cpu().numpy())
    print('실제값(Y): ', Y.cpu().numpy())
    print('정확도(Accuracy): ', accuracy.item())
```
